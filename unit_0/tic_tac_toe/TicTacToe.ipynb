```python
X_CHAR = " x "
O_CHAR = " o "


def print_board():
    header = "   " + "   ".join(str(i) for i in range(len(game_board)))
    print(header)
    for idx, val in enumerate(game_board):
        print(idx, *val)


def check_win():
    for row in game_board:
        if row.count(current_symbol) == len(row):
            print(current_symbol + "win by horizontal strategy")
            return True

    for column in list(zip(*game_board)):
        if column.count(current_symbol) == len(column):
            print(current_symbol + "win by vertical strategy")
            return True

    if [game_board[i][i] for i in range(0, len(game_board))].count(current_symbol) == len(game_board):
        print(current_symbol + "win by diagonal strategy")
        return True

    if [game_board[i][~i] for i in range(0, len(game_board))].count(current_symbol) == len(game_board):
        print(current_symbol + "win by counter diagonal strategy")
        return True

    return False


def ask_dimension():
    while True:
        local_dimension = input("Enter dimension: 3 or 4 ")

        if not (local_dimension.isdigit()):
            print("You need to enter number")
            continue

        local_dimension = int(local_dimension)

        if local_dimension != 3 and local_dimension != 4:
            print("You need to choose 3 or 4 according to the rules")
            continue

        return local_dimension


def ask_coordinates():
    while True:
        cords = input("It's turn for" + current_symbol).split()

        if len(cords) != 2:
            print("Enter two coordinates")
            continue

        local_x, local_y = cords

        if not (local_x.isdigit()) or not (local_y.isdigit()):
            print("You need to enter numbers")
            continue

        local_x, local_y = int(local_x), int(local_y)

        if 0 > local_x or local_x > dimension - 1 or 0 > local_y or local_y > dimension - 1:
            print("Out of bounds")
            continue

        if game_board[local_x][local_y] != " - ":
            print("This cell is already chosen")
            continue

        return local_x, local_y


def game():
    global current_symbol
    while True:
        print_board()
        if check_win():
            break
        else:
            current_symbol = X_CHAR if current_symbol == O_CHAR else O_CHAR
            x, y = ask_coordinates()
            game_board[x][y] = current_symbol


current_symbol = O_CHAR
dimension = ask_dimension()
game_board = [[' - '] * dimension for _ in range(dimension)]
game()
```
